### Getters y Setters

> GETTER: Obtiene el valor de un atributo privado

> SETTER: Modifica el valor de un atributo privado


```python
class PersonaGS:
    def __init__(self, edad:int) -> None:
        self.edad = edad
        self.__curp = "CURP"
    #Getter
    def get_curp(self) -> str:
        return self.__curp
    #Setter
    def set_curp(self, curp:str) -> None:
        self.__curp = curp

    def set_edad(self, edad:int) -> None:
        if edad < 0:
            raise ValueError("La edad no puede ser negativa muñeco") #Lanza una excepcion
        self.__edad = edad


p = PersonaGS(23)
p.get_curp()
```

```python
p.set_curp("CURP2")
p.get_curp() 
```

```python
try:
    p.set_edad(-1)
except ValueError as e:
    print(e)
```

#### Práctica

```python
class Producto:
    def __init__(self) -> None:
        self.__precio = 100.00

    def get_precio(self) -> float:
        return self.__precio

    def set_precio(self, nuevo_precio:float) -> None:
        if nuevo_precio < 0:
            raise ValueError("El precio no puede ser negativo papi.")
        self.__precio = nuevo_precio
```

```python
fifo = Producto()
fifo.get_precio()
```

```python
fifo.set_precio(7.59)
fifo.get_precio()
```

```python
try:
    fifo.set_precio(-1)
except ValueError as e:
    print(e)
```

```python
class Temperatura:
    def __init__(self, celcius:float) -> None:
        self.__celsius = celcius
    
    def get_celcius(self) -> float:
        return self.__celsius
    
    def set_celsius(self, nuevo:float) -> None:
        self.__celsius = nuevo

    def get_fahrenheit(self) -> float:
        return (self.__celsius * 9/5) + 32
```

```python
tempe = Temperatura(50.5)
tempe.get_celcius()
```

```python
tempe.get_fahrenheit()
```

```python
tempe.set_celsius(-50.5)
print(tempe.get_celcius(), tempe.get_fahrenheit())
```

### Decoradores

> Envoltura para una función

```python
def mi_decorador(function):
    def wrapper():
        print("Hola vv antes de la función")
        function()
        print("Adiós vv después de llamar la función")
    return wrapper
```

```python
@mi_decorador
def saludo():
    print("como estás? dentro de la función")

saludo()
```

### Práctica

```python
def nombre_decorador(function):
    def wrapper():
        print("Ejecutando función...")
        function()
    return wrapper


@nombre_decorador
def saludar_nombre():
    print("Hola Rafa")

saludar_nombre()
```

```python
def numero_decorador(function):
    def filtro(numero):
        if numero > 0:
            return function(numero)
        else:
            print("Numero invalido")
    return filtro

@numero_decorador
def raiz(num:float):
    return num**0.5

raiz(25.0)  
```

### @property y @name.setter
> Convertir un método en un atributo

```python
class PersonaD:
    def __init__(self, nombre:str, edad:int)-> None:
        self.nombre = nombre
        self.__edad = edad
    
    @property
    def edad(self) -> int:
        return self.__edad

    @edad.setter
    def nueva_edad(self, edad_nueva:int) -> int:
        if edad_nueva < 0:
            raise ValueError("La edad no puede ser negativa")
        self.__edad = edad_nueva


p = PersonaD("Pepe",50)

p.edad
```

```python
p.nueva_edad = 45
p.edad
```

### Práctica

```python
import math

class Circulo:
    def __init__(self, radio:float)-> None:
        self.__radio = radio
    
    @property
    def radio(self) -> float:
        return self.__radio
    
    @property
    def area(self) -> float:
        return math.pi * self.__radio**2

    @radio.setter
    def nuevo_radio(self, nuevo:float) -> float:
        if nuevo < 0:
            raise ValueError("Pon un radio positivo papi")
        self.__radio = nuevo
```

```python
circ = Circulo(1)
```

```python
circ.radio
```

```python
circ.area
```

```python
circ.nuevo_radio = 2
circ.radio
```
